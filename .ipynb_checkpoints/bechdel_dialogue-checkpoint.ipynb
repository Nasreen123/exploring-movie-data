# Prepare the data

```python
import pandas as pd
```

```python
#Import the data
movie_data = pd.read_csv('movieinfo.csv')
```

```python
movie_data.head()
```

```python
#Improve the column headings to be easier to understand
movie_data = movie_data.rename(columns={'f_dialogue': 'female % main character dialogue', 'm_dialogue': 'male % main character dialogue'})
```

```python
movie_data.head()
```

# Overview of the data

```python
movie_data.describe()
```

We have 2000 movies in this data set, released between 1929 and 2015.  Three quarters of the movies were released since 1992, half since 2001, and a quarter since 2009; so most of them are quite recent.

The average percentage of main characters dialogue that is spoken by female characters is 25.25%.

# Compare movies that pass the bechdel test with those that fail

We will seperate the movies into movies that pass the bechdel test and movies that do not pass the bechdel test.

Will look at how the bechdel test relates to the gender balance of the dialogue, and the year the movie was released.  We will also see if we are more likely to have bechdel data for certain movies, according to these two variables.

For information on the bechdel test: http://bechdeltest.com/

```python
movie_data.groupby('bechdel')['female % main character dialogue', 'year'].describe()
```

We can see that movies that pass the bechdel test have significantly more dialogue by female characters. The average for movies that pass is still below 50%.

There is only a slight different in year of release, with the failing movies released an average of two years earlier.  

```python
pass_bechdel = movie_data[movie_data.bechdel == True]
len(pass_bechdel)
```

fail_bechdel = movie_data[movie_data.bechdel == False]
len(fail_bechdel)

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
plt.figure();
pass_bechdel['female % main character dialogue'].plot.hist(alpha=0.6)
plt.suptitle('% Female characters dialogue for movies that pass the bechdel test')
plt.show()
```

```python
plt.figure();
fail_bechdel['female % main character dialogue'].plot.hist(alpha=0.6)
plt.suptitle('% Female characters dialogue for movies that don\'t pass the bechdel test')
plt.show()
```

```python
ax = pass_bechdel.plot(kind='scatter', x='year',
                  y='female % main character dialogue',
                  color='DarkGreen', label='Pass bechdel')
fail_bechdel.plot(kind='scatter', x='year',
                  y='female % main character dialogue',
                  color='DarkRed', label='Fail bechdel', ax=ax)
plt.show()
```

# Comparing known and unknown bechdel results

```python
unknown_bechdel = movie_data[movie_data.bechdel.isnull()]
len(unknown_bechdel), len(known_bechdel)
```

Almost a third (625) of the movies in out data set have no bechdel result.

```python
unknown_bechdel.describe()
```

```python
known_bechdel = movie_data[movie_data.bechdel.notnull()]
known_bechdel.describe()
```

```python
ax = known_bechdel.plot(kind='scatter', x='year',
                  y='female % main character dialogue',
                  color='DarkOrange', label='Bechdel known')
unknown_bechdel.plot(kind='scatter', x='year',
                  y='female % main character dialogue',
                  color='DarkBlue', label='Bechdel unknown', ax=ax)
plt.show()
```
